```python
import random
import matplotlib.pyplot as plt

width = 10
length = 10

def insidewindow(x, y):
    return 0 <= x < width and 0 <= y < length

def obstacle(x, y):
    if (x, y) in [(0, 3), (1, 3), (2, 3), (3, 3), (4, 3), (7, 3), (8, 3), (9, 3)]:
        return False
    return True

PI = [0.6, 0.2, 0.2]  # probability of moving up, right, and left

def simulate():
    prev_trajectory = [0] * 100
    current_trajectory = [0] * 100
    count_move_prev = 100
    best_path = []

    for k in range(1000):
        state_x, state_y = 0, 0
        count_move = 0
        path = [(state_x, state_y)]

        for l in range(100):
            current_trajectory[l] = 0

        for i in range(100):
            p = random.uniform(0, 1)
            if p <= PI[0]:
                state_y += 1  # MOVE UP
                if insidewindow(state_x, state_y) and obstacle(state_x, state_y):
                    current_trajectory[count_move] = 1
                    count_move += 1
                    path.append((state_x, state_y))
                else:
                    state_y -= 1
            elif p <= PI[0] + PI[1]:
                state_x += 1  # MOVE RIGHT
                if insidewindow(state_x, state_y) and obstacle(state_x, state_y):
                    current_trajectory[count_move] = 2
                    count_move += 1
                    path.append((state_x, state_y))
                else:
                    state_x -= 1
            else:
                state_x -= 1  # MOVE LEFT
                if insidewindow(state_x, state_y) and obstacle(state_x, state_y):
                    current_trajectory[count_move] = 3
                    count_move += 1
                    path.append((state_x, state_y))
                else:
                    state_x += 1

            if state_x == width - 1 and state_y == width - 1:
                print('iteration', k, 'move', count_move, 'goal reached')
                break

        if count_move < count_move_prev:
            count_move_prev = count_move
            prev_trajectory = current_trajectory[:]
            best_path = path[:]

        if k % 100 == 0:
            print('iter', k, 'result', count_move_prev)

    for i in range(100):
        if prev_trajectory[i] != 0:
            print('index', i, 'movement', prev_trajectory[i])

    return best_path

def plot_trajectory(best_path):
    fig, ax = plt.subplots()
    ax.set_xlim(-0.5, width - 0.5)
    ax.set_ylim(-0.5, length - 0.5)
    ax.set_xticks(range(width))
    ax.set_yticks(range(length))
    ax.set_aspect('equal')
    ax.grid(True)
    #ax.axis('off')

    x, y = zip(*best_path)
    ax.plot(x, y, color='red', linestyle='-', marker='o', markersize=4)
    #ax.legend()
    plt.title("Robot Movement Trajectory")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    #plt.axis('off')
    #plt.grid()
    plt.show()

best_path = simulate()
plot_trajectory(best_path)
```
